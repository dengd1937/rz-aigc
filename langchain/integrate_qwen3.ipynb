```python
from dotenv import load_dotenv

load_dotenv()
messages = [
    {
        "role": "user",
        "content": "strawberry有几个r？"
    }
]
```

```python
from langchain_community.chat_models.tongyi import ChatTongyi
chat = ChatTongyi(
    model="qwen2.5-72b-instruct"
)
res = chat.invoke(messages)
res
```

```python
print(res.content)
```

```python
chat = ChatTongyi(
    model="qwen3-235b-a22b",
    model_kwargs={
        "enable_thinking": False
    }
)
res = chat.invoke(messages)
res
```

```python
print(res.content)
```

```python
chat = ChatTongyi(
    model="qwen3-235b-a22b",
    model_kwargs={
        "enable_thinking": True
    },
    streaming=True
)
res = chat.stream(messages)
is_answering = False
print("\n" + "=" * 20 + "思考过程" + "=" * 20 + "\n")
for r in res:
    try:
        if r.chunk_position == 'last':
            break
        if r.additional_kwargs['reasoning_content'] is not None:
            if not is_answering:
                print(r.additional_kwargs['reasoning_content'], end="", flush=True)

        if r.content:
            if not is_answering:
                print("\n" + "=" * 20 + "完整回复" + "=" * 20 + "\n")
                is_answering = True
            print(r.content, end="", flush=True)
    except Exception as e:
        print(r)
```

```python
import base64
from io import BytesIO

def convert_to_base64(pil_image):
    """
    Convert PIL images to Base64 encoded strings

    :param pil_image: PIL image
    :return: Re-sized Base64 string
    """

    buffered = BytesIO()
    pil_image.save(buffered, format="PNG")  # You can change the format if needed
    img_str = base64.b64encode(buffered.getvalue()).decode("utf-8")
    return img_str
```

```python
from PIL import Image

file_path = "image_test.png"
pil_image = Image.open(file_path)

image_b64 = convert_to_base64(pil_image)
image_b64
```

```python
from langchain_core.messages import HumanMessage

def prompt_func(data):
    text = data["text"]
    image = data["image"]

    image_part = {
        "type": "image_url",
        "image_url": f"data:image/jpeg;base64,{image}",
    }

    content_parts = []

    text_part = {"type": "text", "text": text}

    content_parts.append(image_part)
    content_parts.append(text_part)

    return [HumanMessage(content=content_parts)]
```

```python
from langchain_ollama import ChatOllama
from langchain_core.output_parsers import StrOutputParser

llm = ChatOllama(model="qwen3-vl:4b")
chain = prompt_func | llm | StrOutputParser()

query_chain = chain.invoke(
    {"text": "请识别图中的文本内容，言简意赅。", "image": image_b64}
)

print(query_chain)
```
